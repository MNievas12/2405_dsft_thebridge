## Ejercicio 23
Consigue en un DataFrame de todas las cervezas en cuyas cervecerías al menos haya una cerveza con un `abv` mínimo 0.07. Piénsalo bien.

```python
breweries_list = df_all[df_all['abv'] >= 0.07]['brewery_id'].unique()
breweries_list
```

```python
df_all[df_all['brewery_id'].isin(breweries_list)]
```

```python
df_all[df_all['brewery_id'].isin(df_all.groupby('brewery_id')['abv'].max()[df_all.groupby('brewery_id')['abv'].max() >= 0.07].index)]
```

```python
# Obtengo el valor máximo por cervecería
brews = df_all.groupby('brewery_id')['abv'].max()
# Filtro cervecerías que cumplan la condición
brews_filter = brews[brews >= 0.07].index
# Filtro las cervezs que se encuentran en las cervecerías que cumplían la condición
df_all[df_all['brewery_id'].isin(brews_filter)]
```

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media. Piénsalo bien.

```python
media_cervezas = df_all['ibu'].mean()
media_cervezas
```

```python
media_estados = df_all.groupby('state')['ibu'].mean()
lista_estados = media_estados[media_estados > media_cervezas].index
df_all[df_all['state'].isin(lista_estados)]
```

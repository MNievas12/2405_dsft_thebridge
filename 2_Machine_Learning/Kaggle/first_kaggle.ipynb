```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
X = df[['Inches']]
y = df['Price_euros']
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
test = pd.read_csv('data/test.csv')
test.head()
```

```python
y_pred = lr.predict(test[['Inches']])
y_pred
```

```python
submission = pd.DataFrame({'id':test['id'], 'Price_euros':y_pred})
submission
```

```python
submission.to_csv('data/submission_1.csv', index=False)
```
